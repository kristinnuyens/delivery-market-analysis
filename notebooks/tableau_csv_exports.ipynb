Code to export data from takeaway.db to use in Tableau Public...

```python
import sqlite3
import pandas as pd
import os

# Path to your SQLite .db file
db_path = "/Users/kristinnuyens/Desktop/BeCode_Projects/delivery-market-analysis/databases/takeaway.db"

# Folder to save CSVs
output_folder = "/Users/kristinnuyens/Desktop/BeCode_Projects/delivery-market-analysis/sql_outputs/takeaway/db_csv_export"
os.makedirs(output_folder, exist_ok=True)

# Connect to the SQLite database
conn = sqlite3.connect(db_path)

# Get list of tables
tables = pd.read_sql("SELECT name FROM sqlite_master WHERE type='table';", conn)
table_names = tables['name'].tolist()

# Export each table to CSV
for table in table_names:
    df = pd.read_sql(f"SELECT * FROM {table};", conn)
    csv_path = os.path.join(output_folder, f"{table}.csv")
    df.to_csv(csv_path, index=False)
    print(f"Exported {table} → {csv_path}")

conn.close()
print("All tables exported.")
```
